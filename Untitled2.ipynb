**Create a model based on the k-Nearest Neighbours algorithm for classification
based on at least 2 suitable features for one target variable. Verify the optimal k
for classification.**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# ASSUMPTION: The DataFrame 'df' is already loaded.

# --- FIX: Clean Column Names to match the feature list ---
df.columns = df.columns.str.replace(r'[()]', '', regex=True).str.replace(' ', '_')

# --- Feature and Target Selection ---
# Features: Temperature and Luminosity (using the newly cleaned names)
features = ['Temperature_K', 'Luminosity_L/Lo']
# Target: Star Category
target = 'Star_category'

X = df[features].copy()
y = df[target]

# --- 1. Data Preparation: Scaling and Splitting ---

# Scaling: Essential for distance-based algorithms like KNN
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=features)

# Split data: 70% training, 30% testing, stratified to keep class balance
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)

# --- 2. Finding the Optimal K ---
# Test K values from 1 to 20
k_range = range(1, 21)
error_rates = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    # Calculate error rate (1 - accuracy)
    error_rate = 1 - accuracy_score(y_test, y_pred)
    error_rates.append(error_rate)

# Find the optimal k that minimizes the error rate
optimal_k = k_range[np.argmin(error_rates)]
min_error = np.min(error_rates)

print("---------------------------------------------------------")
print(f"Optimal K found: {optimal_k} with minimum error rate: {min_error:.4f}")
print("---------------------------------------------------------")

# --- 3. Visualization for Optimal K ---
plt.figure(figsize=(10, 6))
plt.plot(k_range, error_rates, marker='o', linestyle='--', color='b', markerfacecolor='r', markersize=6)
plt.plot(optimal_k, min_error, marker='o', color='g', markersize=10, label=f'Optimal k={optimal_k}')
plt.title('Error Rate vs. K Value for KNN Classification', fontsize=15)
plt.xlabel('Number of Neighbors (K)', fontsize=12)
plt.ylabel('Error Rate (1 - Accuracy)', fontsize=12)
plt.xticks(k_range)
plt.legend()
plt.grid(True)
plt.show()

# --- 4. Train Final Model and Verify Performance ---
final_knn = KNeighborsClassifier(n_neighbors=optimal_k)
final_knn.fit(X_train, y_train)
y_test_pred = final_knn.predict(X_test)
test_accuracy = accuracy_score(y_test, y_test_pred)

print(f"\nFinal Model (k={optimal_k}) Test Accuracy: {test_accuracy:.4f}")
```

**. Create another eager learning classifier as assigned based on at least 2 suitable
features for one target variable. **

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# ASSUMPTION: The DataFrame 'df' is already loaded.
# --- FIX: Clean Column Names to ensure ML code works ---
df.columns = df.columns.str.replace(r'[()]', '', regex=True).str.replace(' ', '_')

# --- Feature and Target Selection ---
features = ['Temperature_K', 'Luminosity_L/Lo']
target = 'Star_category'

X = df[features].copy()
y = df[target]

# --- 1. Data Splitting ---
# Decision Trees do not require feature scaling.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# --- 2. Create and Train Decision Tree Classifier (Eager Learner) ---
# Limiting max_depth to 5 to avoid excessive overfitting
dt_classifier = DecisionTreeClassifier(random_state=42, max_depth=5)
dt_classifier.fit(X_train, y_train)

# --- 3. Evaluate Model Performance ---
y_test_pred = dt_classifier.predict(X_test)
test_accuracy = accuracy_score(y_test, y_test_pred)

# --- 4. Extract Feature Importance ---
feature_importance = pd.Series(
    dt_classifier.feature_importances_, index=features
).sort_values(ascending=False)

print(f"Decision Tree Classifier Test Accuracy: {test_accuracy:.4f}")
print("\nFeature Importance:")
print(feature_importance.to_markdown(numalign="left", stralign="left", floatfmt=".4f"))
```

**For the eager learner created in Q4, create another model with tuned
hyperparameters that enhance the accuracy of classification.**

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import numpy as np

# ASSUMPTION: The DataFrame 'df' is already loaded and cleaned.
# (Re-running data loading and cleaning internally for tool execution)
df = pd.read_csv("/content/sample_data/Stars.csv")
df.columns = df.columns.str.replace(r'[()]', '', regex=True).str.replace(' ', '_')

# --- Feature and Target Selection and Splitting ---
features = ['Temperature_K', 'Luminosity_L/Lo']
target = 'Star_category'
X = df[features].copy()
y = df[target]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# --- 1. Define Hyperparameter Grid for Tuning ---
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 7, 10, None], # None means unlimited depth
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 3, 5]
}

# --- 2. Use GridSearchCV for Hyperparameter Tuning ---
dt_tuned = DecisionTreeClassifier(random_state=42)
grid_search = GridSearchCV(
    estimator=dt_tuned,
    param_grid=param_grid,
    cv=5,                 # 5-fold cross-validation
    scoring='accuracy',
    n_jobs=-1             # Use all available cores
)

grid_search.fit(X_train, y_train)

# --- 3. Evaluate the Best Model ---
best_dt = grid_search.best_estimator_
y_test_pred = best_dt.predict(X_test)
tuned_accuracy = accuracy_score(y_test, y_test_pred)

print("Best Hyperparameters found by GridSearchCV:")
print(grid_search.best_params_)
print(f"\nTuned Model Test Accuracy: {tuned_accuracy:.4f}")
```
